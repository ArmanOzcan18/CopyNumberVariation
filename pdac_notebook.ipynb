```python
full_ad = sc.read('pdac.h5ad')

indices = np.random.permutation(full_ad.n_obs)

# Select the first half of these indices
half_indices = indices[:full_ad.n_obs // 5]

# Slice the AnnData object to keep only the selected observations
ad = full_ad[half_indices]
print(ad.shape)
ad.to_df().head()
```

```python
threshold = 0.5
filtered_cnas = all_cnas[all_cnas.std(axis=1) > threshold].loc[:,ad.obs.index]
```

```python
from pyensembl import EnsemblRelease

data = EnsemblRelease(77) # will take a while to get the genome data installed in your system

filtered_genes = []
for genomic_region in filtered_cnas.index:
    contig = genomic_region.split(':')[0][3:].lstrip('0')
    start, end = genomic_region.split(':')[1].split('-')
    contig = int(contig)
    start = int(start)
    end = int(end)
    gene_names = data.gene_names_at_locus(contig=contig, position=start, end=end)
    added = [itm for itm in gene_names if itm in ad.var_names]
    filtered_genes.extend(added)

print(len(filtered_genes))
filtered_genes = list(set(filtered_genes))
print(len(filtered_genes))
print(filtered_genes[0])
    
```

```python
n_SEACells = ad.n_obs // 50
print(n_SEACells)
build_kernel_on = 'X_pca'
n_waypoint_eigs = 10

model = SEACells.core.SEACells(ad, 
                  build_kernel_on=build_kernel_on, 
                  n_SEACells=n_SEACells, 
                  n_waypoint_eigs=n_waypoint_eigs,
                  convergence_epsilon = 1e-5,
                  use_sparse=True,)
```

```python
model.construct_kernel_matrix()
M = model.kernel_matrix
```

```python
model.fit(min_iter=10, max_iter=150)
```

```python
import random

def shuffle_list_of_lists(list_of_lists):
    # Flatten the list of lists
    flat_list = [item for sublist in list_of_lists for item in sublist]
    
    # Shuffle the flattened list
    random.shuffle(flat_list)
    
    # Recreate the list of lists with original sublist sizes
    result = []
    index = 0
    for sublist in list_of_lists:
        size = len(sublist)
        new_sublist = flat_list[index:index + size]
        result.append(new_sublist)
        index += size
        
    return result

# Example usage

shuffled_seacells = shuffle_list_of_lists(seacells)

""" for i in range(0, len(shuffled_seacells)):
    print(len(shuffled_seacells[i]))
    print(len(seacells[i])) """
```
